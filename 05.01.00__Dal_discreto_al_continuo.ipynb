# Dal discreto al continuo

Finora abbiamo visto come calcolare la probabilità dell'esito di variabili casuali che hanno distribuzioni discrete.

In ogni caso, l'evento casuale viene ricondotto ad **eventi elementari** che sono considerati **equiprobabili**.

Questo consente di applicare la teoria della probabilità classica, secondo la quale la probabilità di un evento è il rapporto tra il numero di casi favorevoli che corrispondono ad un certo evento ed il numero totale di casi.

La probabilità può essere quindi calcolata con un approccio **combinatoriale**, più o meno complesso dal punto di vista matematico.

# Variabili continue

La difficoltà nel definire la probabilità di un evento determinato da una variabile continua sta nell'impossibilità di definire eventi elementari da considerare equiprobabili.

Anche riducento l'insieme dei possibili esiti ad insiemi contenenti un solo punto, $\{x_0\}$, la presenza di un numero infinito e non numerabile di questi insiemi comporta, assumendo la continuità della probabilità $P(\{x_0\})$, che questa probabilità sia nulla, mentre può essere finita la probabilità associata ad un intervallo che contiene infiniti punti.

Si potrebbe pensare di sostituire il concetto di *equiprobabilità* di eventi elementeri all'*uniformità* della distribuzione di probabilità associata ad un intervallo, cosiché ad intervalli che possono essere messi in corrispondenza secondo qualche criterio geometrico possa venire assegnata la stessa probabilità. Si può vedere però che questa definizione non è univoca, e la scelta della *metrica* da usare resta arbitraria. 

Questa difficoltà viene resa evidente dal paradosso di Bertrand, che può essere formulato come segue.

## Paradosso di Bertrand

Si consideri un triangolo equilatero inscritto in una circonferenza. Si estragga in modo casuale, *uniformemente*, una delle possibili corde del cerchio.

Qual è la probabilità che la lunghezza della corda estratta sia maggiore del lato del triangolo inscritto?

Nella figura che segue sono mostati tre modi possibili di estrarre *uniformemente* una corda del cerchio.
<img src="img/bertrand.png">

1. Si prenda un diametro passante per il vertice superiore del triangolo (fig. a sinistra, tratteggiato in nero). Si estragga *uniformemente* un punto su questo diametro, e si costruisca la corda perpendicolare al diametro. Il processo si può ripetere per simmetria per ogni diametro della circonferenza. Chiaramente, la metà delle corde avrà lunghezza maggiore del lato del triangolo inscritto, quindi $P=1/2$.
2. Si consideri il vertice superiore, e si estragga *uniformemente* un angolo rispetto alla tangente al cerchio passante per quel punto. Il procedimento si può ripetere per qualsiasi punto sulla circonferenza allo stesso modo, per simmetria. In questo caso, un terzo delle corde avrà lunghezza maggiore del lato del triangolo inscritto, quindi $P=1/3$.
3. Si consideri la circonferenza inscritta nel triangolo. Si consideri un punto estratto uniformemente in questo cerchio e si costruisca la corda passante per questo punto, perpendicoare al raggio che passa per lo stesso punto. Tutte queste corde avranno lunghezza maggiore del lato del triangolo inscritto. Se si estrae un punto fuori da questo cerchio e si costruisce una corda allo stesso modo, invece, la lunghezza della corda risulta minore del lato del triangolo inscritto. Il rapporto tra le aree interna ed esterna al cercho è 1/4 perché il raggio interno è la metà di quello esterno, quindi $P=1/4$.

Il paradosso ci dimostra che esistono almeno tre modi diversi per estrarre *uniformemente* una corda, e quindi la variabile casuale che misura la lunghezza della corda.

## Trasformazioni tra variabili continue 

Più in generale, se una variabile casuale $x$ viene trasformata in un'altra variabile $y=f(x)$ attraverso una funzione $f:x\mapsto y=f(x)$ monotona, una distribuzione uniforme per $x$ non corrisponde ad una distribuzione uniforme per $y$.

Questo non è un problema per trasformazioni di variabili discrete dove un valore puntuale viene trasformato in un altro punto da una trasformazione monotona.

**Esempio**: se l'estrazione di un dado mi dà un numero $n$ da 1 a 6, la trasformazione $m=n^2$ mi dà un quadrato perfetto da 1 a 36. Ciascuno dei valori possibili di $m$ avrà probabilità 1/6, uguale alla probabilità di ciascuno dei valori di $n$, mentre i numeri che non sono quadrato perfetto avranno naturalmente probabilità zero, perché non appartengono al *codominio* della funzione $f: n \mapsto m=n^2$.

Negli esempi seguenti mostreremo graficamente come si distribuiscono valori estratti uniformemente o il loro quadrato nel caso discreto e in quello continuo. Il grafico che segue mostra come cambia la spaziatura da 6 valori interi equispaziati ai loro quadrati, che non sono equispaziati ma restano valori individuali puntiformi.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
step = 0.1
x = [np.arange(1, 2+step, step)]*6
y = [(i+1)**x[i] for i in range(6)]
c = [(r, 0, 1-r) for r in np.arange(0, 1, 1/6)]
for i in range(6):
    plt.plot(x[i], y[i], color=c[i])
plt.xticks([1, 2], ['$x$', '$y=x^2$'])
plt.show()
```

```python
plt.figure(figsize=(12,8))
x = [1, 2, 3, 4, 5, 6]
y = [1/6]*6
plt.bar(x, y)
plt.show()
```

```python
plt.figure(figsize=(12,8))
x = np.array([1, 2, 3, 4, 5, 6])**2
y = [1/6]*6
plt.bar(x, y)
plt.show()
```

```python
import numpy.random as random
```

```python
plt.figure(figsize=(12,8))
x = random.uniform(1,6,100000)
n, bins, patches = plt.hist(x, 50, density=True)
```

```python
plt.figure(figsize=(12,8))
y=x**2
n, bins, patches = plt.hist(y, 50, density=True)
```

Sopra, implicitamente, abbiamo confuso la frequenza di un'estrazione casuale con la probabilità, trascurando quella che è la fluttuazione statistica dovuta al fatto che abbiamo un numero di estrazioni finito, assumendo che, con un numero infinito di estrazioni, le fluttuazioni si sarebbero smorzate. Vedremo che questa sarà la base per l'approccio *frequentista* alla probabilità.

In questo senso: se la distribuzione di $x$ è *uniforme*, nel senso che intervalli uguali hanno un ugual numero di estrazioni casuali, $y=x^2$ non avrà la stessa proprietà.

Dobbiamo a questo punto definire meglio cosa significa distribuzione di probabilità nel caso continuo, estendendo quanto abbiamo definito nel caso discreto.
